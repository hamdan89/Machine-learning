```python
# drop rows with data missing for MTSucceed
dataset= dataset.dropna(subset=['MTTotalScore','MTSucceed'],how='any')
```

```python
# pandas drop a column with drop function
dataset= dataset.drop(['idStudent','HW1ScoreVariance','HW1MeanOfDaysBeforeDeadlineOfSubmissions',
                       'HW1VarianceOfDaysBeforeDeadlineOfSubmissions','HW2ScoreVariance','HW2MeanOfDaysBeforeDeadlineOfSubmissions',
                       'HW2VarianceOfDaysBeforeDeadlineOfSubmissions','HW3ScoreVariance','HW3MeanOfDaysBeforeDeadlineOfSubmissions',
                       'HW3VarianceOfDaysBeforeDeadlineOfSubmissions'], axis=1)

```

```python
#kviz20.1                                  97.020484
#kviz20.2                                  95.903166

# pandas drop a column with drop function
dataset= dataset.drop(['kviz20.1','kviz20.2'], axis=1)
```

```python

print(dataset['HW1DateOfFirstSubmission'])
```

```python
#Convert dates to date frame
dataset['HW1DateOfFirstSubmission'] = pd.to_datetime(dataset.HW1DateOfFirstSubmission)
dataset['HW1DateOfLastSubmission'] = pd.to_datetime(dataset.HW1DateOfLastSubmission)
dataset['HW2DateOfFirstSubmission'] = pd.to_datetime(dataset.HW2DateOfFirstSubmission)
dataset['HW2DateOfLastSubmission'] = pd.to_datetime(dataset.HW2DateOfLastSubmission)
dataset['HW3DateOfFirstSubmission'] = pd.to_datetime(dataset.HW3DateOfFirstSubmission)
dataset['HW3DateOfLastSubmission'] = pd.to_datetime(dataset.HW3DateOfLastSubmission)

dataset['HW1DateOfFirstSubmission'] = dataset['HW1DateOfFirstSubmission'].values.astype('float')
dataset['HW1DateOfLastSubmission'] = dataset['HW1DateOfLastSubmission'].values.astype('float')
dataset['HW2DateOfFirstSubmission'] = dataset['HW2DateOfFirstSubmission'].values.astype('float')
dataset['HW2DateOfLastSubmission'] = dataset['HW2DateOfLastSubmission'].values.astype('float')
dataset['HW3DateOfFirstSubmission'] = dataset['HW3DateOfFirstSubmission'].values.astype('float')
dataset['HW3DateOfLastSubmission'] = dataset['HW3DateOfLastSubmission'].values.astype('float')
```

```python


train_test  = dataset.loc[dataset['InValidationSet']  == 0]
validation = dataset.loc[dataset['InValidationSet'] > 0 ]
print('train_test shape is ' , train_test.shape)
print('validation shape is ' , validation.shape)



X_train_test = train_test.iloc[:, :-1].values
Y_train_test = train_test.iloc[:,-1].values

                   
print('X_train_test shape is ' , X_train_test.shape)
print('Y_train_test shape is ' , Y_train_test.shape)


X_validation = validation.iloc[:, :-1].values
Y_validation = validation.iloc[:,-1].values

                   
print('X_validation shape is ' , X_validation.shape)
print('Y_validation shape is ' , Y_validation.shape)


```

```python
# Cleaning data


ImputedModule = SimpleImputer(missing_values = np.nan, strategy ='mean')
ImputedX = ImputedModule.fit(X_train_test)
X_train_test  = ImputedX.transform(X_train_test )

ImputedModule = SimpleImputer(missing_values = np.nan, strategy ='mean')
ImputedX = ImputedModule.fit(X_validation)
X_validation  = ImputedX.transform(X_validation )



print('X_train_test shape is ' , X_train_test.shape)
print('Y_train_test shape is ' , Y_train_test .shape)
```

```python
print(X_train_test)
```

```python

#Splitting data


X_train, X_test, Y_train, Y_test = train_test_split(X_train_test, Y_train_test)


#Splitted Data
print('X_train shape is ' , X_train.shape)
print('X_test shape is ' , X_test.shape)
print('Y_train shape is ' , Y_train.shape)
print('Y_test shape is ' , Y_test.shape)
print('X_validation  shape is ' , X_validation.shape)
print('Y_validation shape is ' , Y_validation .shape)

```

```python
#print (BernoulliNBModel)

BernoulliNBModel = BernoulliNB(alpha=4.3,fit_prior=True)
#BernoulliNB(alpha=1.0, binarize=0.0, fit_prior=True)
BernoulliNBModel.fit(X_train, Y_train)

#Calculating Details
print('BernoulliNBModel Train Score is : ' , BernoulliNBModel.score(X_train, Y_train))
print('BernoulliNBModel Test Score is : ' , BernoulliNBModel.score(X_test, Y_test))
```

```python
pred =BernoulliNBModel.fit(X_train, Y_train).predict(X_validation)
print(pred.tolist())
#print the accuracy score of the model
print("Naive-Bayes accuracy : ",accuracy_score(Y_validation , pred, normalize = True))
```
